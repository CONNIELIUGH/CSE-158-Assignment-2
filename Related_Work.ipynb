**Related Work**

According to the paper Generating and Personalizing Bundle Recommendations on Steam, the authors use the same Steam dataset as our project, but they also include the bundle dataset (which records which bundles of games users bought). Their main goal is to generate and recommend personalized bundles of games for each user, where a bundle is a package of games sold together at a discounted price. They first train an item-level BPR model (Bayesian Personalized Ranking) on user–item interactions, and then build bundle-level models on top of those item embeddings to score and rank bundles.


For bundle recommendation, they propose several models: BR (which basically uses the average user–item scores inside the bundle), BR + N (BR plus an extra term that learns a preference over bundle size), and BR + N + C (BR + N plus a term that encourages bundles whose items are mutually compatible in latent space). They train these bundle models with a BPR objective on user–bundle interactions and evaluate them using performance metrix AUC, which can be interpreted as the probability that a randomly chosen positive example (purchased bundle) gets a higher score than a randomly chosen negative example (non-purchased bundle). Among these, BR + N + C achieves the highest AUC.
The models use a greedy local search algorithm to generate new personalized bundles: starting from a randomly chosen bundle, they repeatedly add, delete, or substitute items from the catalog and keep the bundle that scores best for the user under their model. They evaluate these generated bundles using (1) the average rank of the generated bundle among all existing + generated bundles, and (2) aggregate diversity, to make sure the generated bundles cover a diverse set of items rather than always picking the same popular games.




Even though the paper uses the same underlying Steam data, our project focuses only on the user–item interaction dataset and we do not use the bundle dataset at all. Our task is different: we are predicting whether a user will purchase a game, and if purchased, whether they will actually play it (we noticed a common pattern that many purchased games have playtime equal to zero). Because their work is about bundle recommendation and bundle generation, while ours is about purchase/play prediction, we couldn’t directly compare numerical results like AUC or accuracy.






However, the paper still influenced how we think about our model. Their comparison of BR vs BR + N vs BR + N + C shows that adding meaningful, derived features—like bundle size or item compatibility—can significantly improve performance. That gave us inspiration to be more thoughtful with our own feature engineering, carefully designing and experimenting with features from the user–item interaction data (for example, popularity/engagement signals or compatibility-style features) to get a stronger model for predicting purchase and play behavior.
